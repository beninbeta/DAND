```python
# imports and load data
import pandas as pd
sd =pd.read_csv('store_data.csv', parse_dates = ['week'])
sd.set_index('week', inplace=True)
sdm = sd.resample('M').mean()
```

```python
# explore data
sdm.describe()
```

```python
# sales for the last month

sd_bar = sdm.tail(1).plot(kind='bar', figsize=(10,6))
sd_bar.set_ylabel("Sales", fontsize=20)
```

```python
# average sales
sd_avg = sd.mean()
sd_avg.plot(kind='pie', figsize=(8,8))
```

```python
# sales for the week of March 13th, 2016
sd_mar = sd[sd.index == '2016-03-13']
sd_mar.plot(kind='bar', figsize=(8,8))
```

```python
# sales for the lastest 3-month periods
sd_lastq = sdm.tail(3).sum().plot(kind='bar', figsize=(8,8))
```
